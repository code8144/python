```python
csv_data =[]
with open('전국건강증진센터표준데이터1.csv') as f:
    for line in f:
        csv_data.append(line.split(','))
print(csv_data[:3])
```

```python
len_list =[]
for e in csv_data:
    len_list.append(len(e))
print(set(len_list))
```

```python
doctor_index = csv_data[0].index('의사수')
print(doctor_index)
```

```python
for i,e in enumerate(csv_data[0]):
    print(i,e)
```

```python
doctor_count = []
for e in csv_data[1:]:
    doctor_count.append(e[doctor_index])
#print(doctor_count)
print('center count : ', len(doctor_count))
print('center with no doctor : ', doctor_count.count('0'))
```

```python
#print(doctor_count)
for i, c in enumerate (doctor_count):
    try:
        doctor_count[i] = int(c)
    except:
        doctor_count[i] = 0
print('total doctor count : ', sum(doctor_count))
```

```python
name_list=[]
for e in csv_data[1:]:
    if len(e[2])>0:
        name_list.append(e[2].split()[0])
        #print(e[2].split()[0])
print(name_list)
```

```python
for n in name_list:
    if n[0]=='"':
        print(n[1:])
    else:
        print(n)
```

```python
for i, n in enumerate(name_list):
    name_list[i] = n.lstrip('"')
    #print(n.lstrip('"'))
print(name_list[:10])
```

```python
#name_set = set(name_list)
for n in set(name_list):
    print(n, name_list.count(n))
```

```python
def sort_key(x):
    return x[1]

def unique_count(raw_list, sort = False):
    raw_set = set(raw_list)
    v_list = []
    for n in raw_set:
        v_list.append(raw_list.count(n))

    ret_dict = dict(zip(raw_set, v_list))
    if sort == True:
        #ret_dict = dict(sorted(ret_dict.items(), key = lambda x: x[1]))
        ret_dict = dict(sorted(ret_dict.items(), key = sort_key))
    return ret_dict  
    
unique_count(name_list, sort = True)
```
